```python
import pandas as pd
data_frame=pd.read_csv('data.csv')
data_frame.shape
```

```python
df1=pd.DataFrame(data_frame,columns=['original_title','budget','revenue'])
df1['benefit']=df1['revenue']-df1['budget']
df1.sort_values('benefit',ascending=False)
```

```python
import seaborn as sns
sns.set()
graph=sns.scatterplot(x='budget',y='revenue',data=df1)
graph
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool

TOOLTIPS=HoverTool(tooltips=[
    ("index","$index"),
    ('(budget,revenue)',"(@budget , @revenue)"),
    ("original_title","@original_title"),
])

p = figure(title="filmbaz graph", x_axis_label="budget", y_axis_label="revenue",tools=[TOOLTIPS ])
p.circle('budget','revenue',size=10,source=df1)
show(p)
```
